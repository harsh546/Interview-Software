```python
from pydub import AudioSegment

def split_audio_file(file_path, chunk_length_in_seconds):
    
    audio = AudioSegment.from_wav(file_path)
    chunks = []
    
    for i in range(0, len(audio), chunk_length_in_seconds * 1000):
        
        chunk = audio[i:i + chunk_length_in_seconds * 1000]
        chunks.append(chunk)
        
    return chunks
```

```python
def save_chunks_as_wav_files(chunks, output_folder):
    
    for i, chunk in enumerate(chunks):
        
        output_file_path = f"{output_folder}/chunk_{i}.wav"
        chunk.export(output_file_path, format="wav")
```

```python
import speech_recognition as sr
from pydub import AudioSegment

def speech_to_text(file_path):
    
    # Load the audio file
    audio = AudioSegment.from_wav(file_path)

    # Convert the audio file to the correct format for speech recognition
    audio.export("temp.wav", format="wav")

    # Use the speech recognition library to convert the speech in the audio file to text
    recognizer = sr.Recognizer()
    
    with sr.AudioFile("temp.wav") as source:
        audio_data = recognizer.record(source)
        text = recognizer.recognize_google(audio_data)

    return text
```

```python
import nltk
from nltk.corpus import stopwords

def remove_stopwords(word_list):
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in word_list if word not in stop_words]
    return filtered_words
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

def calculate_compound_score(strings):
    sia = SentimentIntensityAnalyzer()
    scores = []
    for string in strings:
        sentiment_score = sia.polarity_scores(string)
        compound_score = sentiment_score['compound']
        scores.append(compound_score)
    return scores
```

```python
def labelling(score,label):
    
    for word in score:
        
        if word >= 0.05:
        
           label.append('pos')
        
        elif word <= -0.05:
        
           label.append('neg')    
            
        else:
        
           label.append('neu')    
```

```python
def count_and_remove_duplicates(input_list):
  """Counts the number of times each string appears in a list and removes duplicates.
  Args:
    input_list: The list to search.
  Returns:
    A dictionary with the strings as keys and the number of times they appear in the list as values.
  """
  counts = {}
  unique_list = []

  for item in input_list:
        
    if item not in counts:
      counts[item] = 0
    counts[item] += 1
    
    if item not in unique_list:
      unique_list.append(item)
    
  return counts, unique_list
```

```python
# Example usage
input_list = ['apple', 'banana', 'cherry', 'apple', 'banana', 'grape', 'apple']
counts, unique_list = count_and_remove_duplicates(input_list)
print(f'The number of times each string appears in the list is: {counts}')
print(f'The list with duplicates removed is: {unique_list}')
```

```python
def get_values(dictionary):
  """Returns a list of the values in a dictionary.
  Args:
    dictionary: The dictionary to get the values from.
  Returns:
    A list of the values in the dictionary.
  """
  values = []
  for key in dictionary:
    values.append(dictionary[key])
  return values
```
